```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
data = pd.read_csv(r"D:\Program Files (x86)\pythonpro\SpringFestivaldata\data\02_pandas\02_Data_preprocessing\boss岗位.csv",
                    encoding="gbk")
data = data.loc[1:,:]
data.head()
```

```python
# 完整性分析
data.info()
```

##### 1 只要标题含“数据分析”的数据，删除标题 地区 公司名 薪资相同的数据

```python
data[data.标题.str.contains("数据分析")]
data.drop_duplicates(subset=["标题","地区","公司名","薪资"],keep="last")
data.describe()
```

##### 2 从地区中提取出城市、区、商圈，并删除地区列

```python
data["城市"] = data.地区.str.extract("(.*?)·")
data["区"] = data.地区.str.extract("·(.*)·")
data["商圈"] = data.地区.str.extract("·(\w*)$")         # \w数字、字母、汉字，不包括特殊字符
del data["地区"]
data.head()
```

##### 6 从薪资中提取薪资下限、薪资上限、薪数，并删除薪资列

```python
data["薪资下限(K)"] = data.薪资.str.extract("(.*?)-").astype(int)
data["薪资上限(K)"] = data.薪资.str.extract("-(.*)K").astype(int)
data["薪数"] = data.薪资.str.extract("·(.*)薪")
data.薪数.fillna("12",inplace=True)
del data["薪资"]
data.薪数.astype(int)
data.head()
```

##### 7 北京的平均薪资


```python
(data["薪资下限(K)"] + data["薪资上限(K)"])/2 * (data.薪数.astype(float))/12
# sal = (data["薪资下限(K)"] + data["薪资上限(K)"])/2
# sal.multiply(data.薪数.astype(float))/12
data["平均薪资"] = (data["薪资下限(K)"] + data["薪资上限(K)"])/2 * (data.薪数.astype(float))/12
data[data.城市=="北京"].平均薪资.mean()
# data.平均薪资[data.城市=="北京"].mean()
```

```python
# 统计各城市招聘岗位数，平均薪资，最高薪资，最低薪资，薪资标准差
data.groupby("城市").agg({"标题":"count","平均薪资":["mean","max","min","std"]})
```

```python
# 统计各城市各经验平均薪资（透视表与交差表分别实现）
pd.pivot_table(data,index="城市",columns="经验",values="平均薪资",aggfunc="mean")       # 透视表
pd.crosstab(index=data.城市,columns=data.经验,values=data.平均薪资,aggfunc="mean")
```

```python
# 找到每个城市薪资最高的区
    # 各城市各区的平均薪资
df = data.groupby(["城市","区"],as_index=False)["平均薪资"].mean()
df
    # 城市内以平均薪资给各区排名
df_rank = df.groupby("城市")["平均薪资"]
df["名次"] = df_rank.rank(ascending=False)
df
    # 找到各城市中平均薪资名次为1的区
df[df.名次==1]
```

```python
rank_test = pd.DataFrame({"school":list("aaabbb"),"score":[80,80,70,50,60,60]})
rank_test_rand = rank_test.groupby("school")["score"]
rank_test["average"] = rank_test_rand.rank(method="average",ascending=False)
rank_test["min"] = rank_test_rand.rank(method="min",ascending=False)
rank_test["max"] = rank_test_rand.rank(method="max",ascending=False)
rank_test["first"] = rank_test_rand.rank(method="first",ascending=False)
rank_test["dense"] = rank_test_rand.rank(method="dense",ascending=False)
rank_test
```
